```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from sklearn import linear_model
```

```python
df = pd.read_csv("D:\MLOM\Machine-Learning\Linear_Regression_Model\csv files\canada_per_capita_income.csv")
```

```python
df
```

```python
if 'year\tper capita income (US$)' in df.columns:
    df[['year', 'per capita income (US$)']] = df['year\tper capita income (US$)'].str.split('\t', expand=True)
```

```python
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['per capita income (US$)'] = pd.to_numeric(df['per capita income (US$)'], errors='coerce')
```

```python
df = df.dropna()
```

```python
df.columns
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']], df[['per capita income (US$)']])
```

```python
# future_years = pd.DataFrame({'year': [2017, 2030]})
```

```python

future_years = pd.DataFrame({'year': list(range(2017, 2031))})
future_predictions = reg.predict(future_years)
```

```python
future_predictions
```

```python
future_df = future_years.copy()
future_df['predicted income (US$)'] = future_predictions
```

```python
future_df
```

```python
# df.to_csv("D:\MLOM\csv files\canada_per_capita_income.csv" , index=False)
```

```python
df_extended = pd.concat([df, future_df], ignore_index=True)
```

```python
df_extended.to_csv("yourfile_with_predictions.csv", index=False)
```
