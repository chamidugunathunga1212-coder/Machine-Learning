```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("D:\MLOM\Machine-Learning\Linear_Regression_Model\csv files\homeprices.csv")
```

```python
df
```

```python
%matplotlib inline
plt.scatter(df.area, df.price, color="blue", marker="o")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']], df.price)
```

```python
# Predict multiple values
areas = np.array([3000, 4000, 5000]).reshape(-1, 1)
predicted_prices = reg.predict(areas)
print(predicted_prices)
```

```python
reg.predict(np.array([[5000]]))
```
